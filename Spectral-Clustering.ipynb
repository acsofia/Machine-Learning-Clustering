```python
import numpy as py
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
dataset
```

```python
X=dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import SpectralClustering
sc=SpectralClustering(n_clusters=8, assign_labels='discretize',random_state=0).fit(X)
label=sc.fit_predict(X)
```

```python
supervised=dataset
```

```python
supervised['Cluster_group']=label
supervised
```

```python
import seaborn as sns
facet=sns.lmplot(data=supervised,
                x=supervised.columns[3],
                y=supervised.columns[4],
                hue=supervised.columns[5],
                legend=True,fit_reg=False,legend_out=True)
```
